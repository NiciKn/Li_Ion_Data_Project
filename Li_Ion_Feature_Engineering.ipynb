# NASA randomized and recommissioned battery dataset

An accelerated Life Testing Dataset for Lithium-Ion Batteries with Constant and Variable Loading Conditions:

The dataset repository is organized into three main folders, each containing one group of life cycled battery packs. 
Within each folder individual battery packs own their dedicated csv file for continuous data logging, which are named with their respective battery pack number.

The folders are named: 

- regular_alt_batteries: Containing one csv file for each battery pack cycled at the same load level or load range throughout lifetime
- recommissioned_batteries: Containing one csv file for each battery pack cycled at different load levels at varying life stages
- second_life_batteries: Containing one csv file for each second life battery pack cycled at constant current througout the second life 

The columns in each csv file contain the following data with the provided units: 

The following columns contain data throughout the cycling process: 

- start time: [mm:dd:yyyy hh:mm:ss] (start time of each cycle day for each battery pack, in most cases approximately 24h)
- relative time: [s] (the relative time is continuous from the beginning of the entire life cycle to the failure of each battery pack)
- mode: -1 = discharge, 0 = rest, 1 = charge
- voltage charger: [V] (continuous battery pack voltage measurement right after the connection of the battery pack to the charger board)
- temperature battery: [C] (continuous temperature measurement on the battery cell electrode surface)

Those columns only contain data when the respective battery is connected to the load board and performing a discharge mission

- voltage load: [V] (battery pack voltage measured on the load board)
- current load: [A] (discharge current measured on the load board using current sense resistors)
- temperature mosfet: [C] (temperature on the load board mosfets, measured for safety purposes)
- temperature resistor: [C] (temperature on the load board current sense resistor, measured for safety purposes)
- mission type: 0 = reference discharge (constant current at 2.5A), 1 = regular mission

## Regular accelerated life test (regular alt batteries folder):

This folder contains the data files from battery packs subject to constant and variable loading conditions, 
where the current range of the latter remains the same throughout the battery life.

The assigned battery packs in this folder are numbered as
follows:

– Constant current:
Those battery packs were subjected to the same load level during discharge missions over their respective life time.

The here provided averaged current levels are approximated, and might vary slightly:

* 9.30A: Battery pack 0.1 and 1.1
* 12.9A: Battery pack 3.1 and 2.2
* 14.3A: Battery pack 2.3 and 5.2
* 16.0A: Battery pack 0.0 and 1.0
* 19.0A: Battery pack 2.0, 3.0 and 2.1

– Variable current:

Those battery packs are subjected to variable current within a discharge cycle.
The current level segments are kept constant for either 40, 60 or 80 seconds before switching to another level.
Both, the amplitude and length of each current level segments was randomly selected to create a pool of different random loading discharge missions.

The here provided averaged current levels are approximated, and might vary slightly:

* 14.3A average (12.9A - 16A range): Battery pack 4.1 and 5.1
* 17.0A average (16A - 19A range): Battery pack 4.0 and 5.0


## Recommissioned batteries (recommissioned batteries folder):

This folder contains the data files from battery packs where the constant loading conditions were changed throughout the battery life. 
The assigned battery packs in this folder are numbered as follows:

– Two life stages:
The following batteries were subjected to constant current levels at the first provided amplitudes until reaching an capacity level of approximately 2.2Ah,
and then switched to the second current level until reaching end of life

The here provided averaged current levels are approximated, and might vary slightly:

* 16A, 14.3A: Battery pack 3.2
* 14.3A, 16A: Battery pack 5.3
* 16A, 12.9A: Battery pack 0.2 and 3.3
* 16A, 9.30A: Battery pack 1.2 and 2.4

– Three life stages:
Those batteries were subjected to current level switches at approximately 2.2Ah and 2.0Ah of capacity.
The here provided averaged current levels are approximated, and might vary slightly:

* 16A, 14.3A, 7.5A: Battery pack 0.3
* 16A, 7.5A, 14.3A: Battery pack 2.5


## Second life batteries (second life batteries folder):

This folder contains the data files from battery packs that consist of battery cells survived from previously cycled battery packs. 
The assigned battery packs in this folder are numbered as follows:

– 2nd life battery packs:
The following batteries were subjected to constant current levels at the first provided amplitudes until reaching end of life (EOL).
Survivor cells were then reused and subjected to the second provided current level until reaching EOL

The here provided averaged current levels are approximated, and might vary slightly:

* 16.0A, 9.3A: Battery pack 1.3 (First life battery packs: 0.0 & 1.1)
* 14.3A, 7.5A: Battery pack 3.6 (First life battery packs: 2.3 & 5.2)
* 12.9A, 5.0A: Battery pack 5.4 (First life battery packs: 2.2 & 3.1)

## Reinigen der Daten
* Nicht benötigte Spalten entfernen: start_time, voltage_load, temperature_mosfet, temperature_resistor
* Fehlende Werte ersetzen: mission_type, teilweise temperature_battery
* Unsinnige Werte ersetzen: Temperaturen werden nach unten auf 20 °C begrenzt (selten Ausreißer bis auf -100 °C)
* Unsinnige Werte entfernen: Teilweise 0V im Ladevorgang
* Spalten umbenennen: 'voltage_charger'->'voltage', 'temperature_battery'->'temperature'
* delta_t berechnen


````python
#Clean data and drop not used columns
def clean_data(df_in):
    df = df_in.copy()
    df['temperature_battery'] = df['temperature_battery'].astype(float)

    # fill NA values of mission_type with 1 (normal mission)
    df['mission_type'] = df['mission_type'].fillna(1.0)
    
    # fill NA values of temperature and replace unrealistic values
    mean_temp = df['temperature_battery'].mean()
    df['temperature_battery'] = df['temperature_battery'].fillna(mean_temp)
        
    df['temperature_battery'] = df['temperature_battery'].apply(lambda x: max(x,20))
    df['current_load'] = df['current_load'].fillna(0)
    df = df.drop(columns=['start_time','temperature_mosfet','temperature_resistor','voltage_load'])

    #Rename voltage_charger to voltage, since there is only one voltage left, same with temperature
    df.rename(columns={'voltage_charger':'voltage'}, inplace=True)
    df.rename(columns={'temperature_battery':'temperature'}, inplace=True)
    
    # remove voltage zero values
    df = df[df['voltage'] != 0].copy()
    
    # reset index
    df.reset_index(drop=True, inplace=True)
    
    # delta t berechnen
    df['delta_t'] = df['time'].diff().fillna(0)
    
    return df
````

## Feature Engineering
* Generiert aus zeitbasierten Features zyklusbasierte Features
* Kumulierte Features:
    * time_sec: Betriebsdauer in Sekunden
    * EFC: Äquivalente Vollzyklen
    * E_cum_Wh: Entladene Energie in Wh
    * T_cum_avg: Durchschnittliche kumulierte Temperatur in °C
    * T_cum_max: Maximale kumulierte Temperatur in °C
    * T_cum_min: Minimale kumulierte Temperatur in °C
    * Temperatur Histogramm in Sekunden: 'sec_T_st_-20','sec_T_-20_0', 'sec_T_0_15', 'sec_T_15_30', 'sec_T_30_45', 'sec_T_45_60','sec_T_60_80', 'sec_T_gt_80'
    * C_rate_cum_avg: Durchschnittliche kumulierte C (Entlade-)rate
    * C_rate_cum_max: Maximale kumulierte C (Entlade-)rate
    * C rate Histogramm in Sekunden: 'sec_C_rate_0_0.5', 'sec_C_rate_0.5_1', 'sec_C_rate_1_2', 'sec_C_rate_2_3', 'sec_C_rate_3_4','sec_C_rate_4_5', 'sec_C_rate_5_6', 'sec_C_rate_6_7', 'sec_C_rate_gt_7'
    * sec_at_high_voltage: Sekunden über 2*4,0V
    * sec_at_low_voltage: Sekunden unter 2*3,1V
    * integral_high_volt_high_temp: sum(max(voltage-2*4.0V, 0)*max(temperature-40°C, 0)*delta_t)
    * integral_low_volt_high_temp: sum(max(2*3,1V-voltage, 0)*max(temperature-40°C, 0)*delta_t)
    * integral_low_volt_high_discharge: sum(max(2*3,1V-voltage, 0)*max(current_load-1C, 0)*delta_t)
    * integral_high_temp_high_discharge: sum(max(temperature-40°C, 0)*max(current_load-1C, 0)*delta_t)
    * integral_low_temp_high_discharge: sum(max(0°C-temperature, 0)*max(current_load-1C, 0)*delta_t)
    * integral_discharged_squared: sum(current_load ** 2 * delta_t)
    * integral_discharged_squared_T: sum(current_load ** 2 * (10 + abs(T - 20°C)) * delta_t)
    * sec_T_over_X: Sekunden bei der die Temperatur über X °C liegt
* Zyklusspezifische Features:
    * batt_label: Label der getesteten Batterie
    * isRefCycle: True: Referenzzyklus mit 2.5A Entladestrom, False: Normaler Zyklus
    * Q_norm_at_3.xV: Ladung die benötigt wird, um Batterie von voller Ladung auf 2*3.xV zu entladen normiert auf nominale    Kapazität.Wird nur in den Referenzzyklen berechnet, dazwischen wird linear interpoliert.
    * T_cycle_avg: Durchschnittliche Temperatur des Zyklus in °C
    * T_cycle_max: Maximale Temperatur des Zyklus in °C
    * T_cycle_min: Minimale Temperatur des Zyklus in °C
    * C_rate_cycle_avg: Durchschnittliche C (Entlade-)rate des Zyklus
    * C_rate_cycle_max: Maximale C (Entlade-)rate des Zyklus
    * R_mOhm: Gemessener (DC) Innenwiderstand in mOhm
    * SOH: State of Health (C_zyklus/C_nom) in %. Wird nur in den Referenzzyklen berechnet, dazwischen wird linear interpoliert
    * SOH(-30EFC), SOH(-60EFC), SOH(-90EFC), SOH(-120EFC): Der SOH Wert vor 30, 60, 90, 120 EFC (equivalent full cycles)
    * delta_SOH(-30EFC), delta_SOH(-60EFC), delta_SOH(-90EFC), delta_SOH(-120EFC): Die SOH Änderung seit 30, 60, 90, 120 EFC
    * RUL_EFC: Verbleibende nutzbare Lebensdauer (Remaining useful life) in äquivalenten Vollzyklen

#### Hilfsfunktionen

````python
def calc_Q_V(voltage, current, delta_t, voltage_points, nom_cap_mAh, min_start_voltage):
    '''
    Calculates the normalized discharge from full charge for given voltage points
    '''
    Q_norm = []
    if voltage[0] < min_start_voltage:
        return False, []
    i=0
    for idx in range(len(voltage_points)):
        voltage_point = voltage_points[idx]
        while i < len(voltage)-1 and voltage[i] > voltage_point:        
            i += 1
        q = (current[:i] * delta_t[:i]).sum()
        Q_norm.append(q/3.6/nom_cap_mAh)
    return True, Q_norm

def return_inner_res_mOhm(mode, current, voltage, min_current,min_volt_drop):
    '''
    Calculates the inner resistance when the mode changes from discharge to idle
    '''
    max_delay_sec = 10
    falseRet = (False,0)

    discharge_idx = np.where(current > min_current)[0]

    #When no discharge current greater min_current found return false
    if len(discharge_idx) == 0:
        return falseRet

    start_idx=discharge_idx[-1]

    curr_val = current[start_idx] 
    if curr_val < min_current:
        return falseRet
    
    for offset in range(max_delay_sec):
        idx = start_idx + offset
        voltage1 = voltage[idx]
        voltage2 = voltage[idx+1]
        voltage_drop = voltage2 - voltage1
        if voltage_drop > min_volt_drop:
            res = 1000*voltage_drop/curr_val
            return True,res

    return falseRet

    
    
def isReferenceCycle(mission, current):
    '''
    checks if given cycle is a reference cycle with 2.5A discharge
    '''
    for i in range(len(current)):
        if (mission[i] == 0) and (2 < current[i] < 3):
            return True
    return False

def calc_charge(current, delta_t):
    '''
    calculates the full charge throuput for the given data
    '''
    charge_As = (current*delta_t).sum()
    charge_mAh = charge_As / 3.6
    return charge_mAh

def calc_energy(current, voltage, delta_t):
    '''
    calculates the full energy throuput for the given data
    '''
    energy_Ws = (current * voltage * delta_t).sum()
    energy_Wh = energy_Ws / 3600
    return energy_Wh

def calc_hist(feature, delta_t, bins):
    '''
    Calculates histogram
    '''
    hist = [0]*(len(bins)+1)

    bins = [-np.inf] + bins + [np.inf]
    
    for i in range(0,len(bins)-1):
        bool_idx = (feature >= bins[i]) & (feature < bins[i+1])
        hist[i] = delta_t[bool_idx].sum()
    
    return hist

def calc_average(feature, delta_t, integral, sum_time):
    '''
    Calculates the cycle average, cumulated avarage, the new time and integral values
    '''
    sum_time_cycle = delta_t.sum()
    integral_cycle = (feature * delta_t).sum()

    cycle_avg = integral_cycle / sum_time_cycle

    integral += integral_cycle
    sum_time += sum_time_cycle

    cum_avg = integral / sum_time
    
    return cycle_avg, cum_avg, integral, sum_time

def calc_thres_times(feature, delta_t, thres):
    '''
    Calculate time over thresholds
    '''
    thres_times = [0]*(len(thres))
    for i in range(len(thres)):
        thres_times[i] = delta_t[feature >= thres[i]].sum()

    return thres_times

def integrate_two_feat_prod(delta_t, feature1, operand1, thres1, feature2, operand2, thres2):
    '''
    integrates product of two features over time with given threshold, with operand 'gt' for
    'greater than' or operand 'lt' for 'lower than'
    '''
    features = [feature1, feature2]
    operands = [operand1, operand2]
    thresholds = [thres1, thres2]
    diffs = 2*[0]
    i=0
    for feature, operand, thres in zip(features, operands, thresholds):
        if operand == 'gt':
            diffs[i] = np.maximum(feature-thres, 0)
        elif operand == 'lt':
            diffs[i] = np.maximum(thres-feature, 0)
        else: #invalid operand
            return 0
        i += 1
    integral = (diffs[0] * diffs[1] * delta_t).sum()
    return integral

        
def return_full_charged_cycle_indices(mode, voltage, charged_volt, discharged_volt):
    '''
    return all indices where a new cycle starts
    '''
    indices = []
    max_idx = len(mode) - 1
    i=0
    idx_step = 10
    
    # check time when battery is fully charged
    while i < max_idx:
        if mode[i] == 0 and voltage[i] >= charged_volt:
            break
        i += idx_step
    
    while i < max_idx:
        indices.append(i)
        # wait until battery is discharging
        while (i < max_idx) and (mode[i] != -1):
            i += idx_step
        
        # check time when battery is fully discharged
        while i < max_idx:
            if mode[i] == 0 and voltage[i] <= discharged_volt:
                break         
            i += idx_step

        # check time when battery is fully charged
        while i < max_idx:
            if mode[i] == 0 and voltage[i] >= charged_volt:
                break
            i += idx_step
            
    return indices

def get_soh(soh_vector, efc_vector, num_efc):
    '''
    Get SOH value for given EFC
    '''
    if num_efc <= efc_vector[0]:
        return 100
    else:
        
        max_idx = len(efc_vector)-1
        for i in range(max_idx):
            if efc_vector[i] <= num_efc < efc_vector[i+1]:
                break
        if i >= max_idx:
            #something wrong
            return 100
        else:
            #make linear interpolation when index found
            soh1 = soh_vector[i]
            soh2 = soh_vector[i+1]
            efc1 = efc_vector[i]
            efc2 = efc_vector[i+1]
            soh = soh1 + ((num_efc-efc1)*(soh2-soh1)/(efc2-efc1))
            return soh
            
            
````

#### Generierung der zyklusbasierten Daten

````python
def generate_cycle_based_data(batt_label, data):
    """
    Generate cycle based data from the time based full cycle battery test data
    """   
    d = {}

    # EFC, E_cum_Wh and general stuff
    sum_Q_mAh = 0
    sum_E_Wh = 0
    for feature in ['batt_label', 'isRefCycle', 'time_sec', 'EFC', 'E_cum_Wh']:
        d[feature] = []

    # Variables for Q(V) calculation
    voltage_points = [7.8, 7.6, 7.4, 7.2]
    Q_V_norm_features = ['Q_norm_at_3.9_V', 'Q_norm_at_3.8_V', 'Q_norm_at_3.7_V', 'Q_norm_at_3.6_V']
    for feature in Q_V_norm_features:
        d[feature] = []

    # Variables for temperature histogram and avg,min,max calculation
    temp_features = ['T_cycle_avg', 'T_cycle_max', 'T_cycle_min', 'T_cum_avg', 'T_cum_max', 'T_cum_min']
    for feature in temp_features:
        d[feature] = []

    temp_thres = [30, 40, 45, 50, 60]
    temp_thres_features = ['sec_T_over_30', 'sec_T_over_40', 'sec_T_over_45', 'sec_T_over_50', 'sec_T_over_60']
    for feature in temp_thres_features:
        d[feature] = []
    
    T_hist_features = ['sec_T_st_-20','sec_T_-20_0', 'sec_T_0_15', 'sec_T_15_30', 'sec_T_30_45',\
                       'sec_T_45_60','sec_T_60_80', 'sec_T_gt_80']
    for feature in T_hist_features:
        d[feature] = []
        
    sum_time_temp_thres = np.array(len(temp_thres)*[0.0])
    temp_bins = [-20, 0, 15, 30, 45, 60, 80]
    temp_hist_len = len(temp_bins)+1
    temp_hist = np.array(temp_hist_len*[0.0])
    integral_temp = 0
    max_temp = -1000
    min_temp = 1000
    sum_time = 0

    # Variables for current histogram and avg,max calculation
    sum_discharge_time = 0
    C_rate_features = ['C_rate_cycle_avg', 'C_rate_cycle_max', 'C_rate_cum_avg', 'C_rate_cum_max']
    for feature in C_rate_features:
        d[feature] = []

    C_hist_features = ['sec_C_rate_0_0.5', 'sec_C_rate_0.5_1', 'sec_C_rate_1_2', 'sec_C_rate_2_3',\
                       'sec_C_rate_3_4','sec_C_rate_4_5', 'sec_C_rate_5_6', 'sec_C_rate_6_7', 'sec_C_rate_gt_7']
    for feature in C_hist_features:
        d[feature] = []
        
    curr_bins_C = [0.5,1,2,3,4,5,6,7]
    curr_hist_len = len(curr_bins_C)+1
    curr_hist = np.array(curr_hist_len*[0.0])
    integral_curr_C = 0
    max_curr_C = -1000

    # Variables for ov/uv calculations
    volt_hist_len = 3
    volt_hist = np.array(volt_hist_len*[0.0])
    d['sec_at_high_voltage'] = []
    d['sec_at_low_voltage'] = []

    # Variables for integrals
    integrals = {}
    integral_features = ['integral_high_volt_high_temp', 'integral_low_volt_high_temp', 'integral_low_volt_high_discharge',\
                        'integral_high_temp_high_discharge', 'integral_low_temp_high_discharge', 'integral_discharge_squared',\
                        'integral_discharge_squared_T']
    for feature in integral_features:
        d[feature] = []
        integrals[feature] = 0

    # Variables for R_mOhm and SOH
    res_mOhm = 50
    nom_capacity_mAh = 2450
    d['R_mOhm'] = []
    d['SOH'] = []

    # convert data to numpy arrays to optimize runtime (from 1h to 3 min for generating all data)
    mode_complete = data['mode'].to_numpy()
    mission_complete = data['mission_type'].to_numpy()
    time_complete = data['time'].to_numpy()
    delta_t_complete = data['delta_t'].to_numpy()
    current_complete = data['current_load'].to_numpy()
    temp_complete = data['temperature'].to_numpy()
    voltage_complete = data['voltage'].to_numpy()

    # Get indices for every full cycle at full charge
    ref_indices = []
    indices = return_full_charged_cycle_indices(mode_complete, voltage_complete, 8, 6.5)

    # just a helper variable to fix runaway value at battery51
    batt51_refCycle_cnt = 0
 
    for i in range(0,len(indices)-1):
        d['batt_label'].append(batt_label)
        start_idx = indices[i]
        end_idx = indices[i+1]

        # define data for this cycle
        mode = mode_complete[start_idx:end_idx]
        mission = mission_complete[start_idx:end_idx]
        time = time_complete[start_idx:end_idx]
        delta_t = delta_t_complete[start_idx:end_idx]
        current_A = current_complete[start_idx:end_idx]
        current_C = 1000 * current_A / nom_capacity_mAh
        temp = temp_complete[start_idx:end_idx]
        voltage = voltage_complete[start_idx:end_idx]

        # check if reference cycle
        referenceCycle = isReferenceCycle(mission, current_A)
        d['isRefCycle'].append(referenceCycle)

        # append last time value
        d['time_sec'].append(time[-1])

        # calc charge and energy for this cycle
        charge_mAh = calc_charge(current_A, delta_t)
        energy_Wh = calc_energy(current_A, voltage, delta_t)

        # accumulate total charge and energy
        sum_Q_mAh += charge_mAh
        sum_E_Wh += energy_Wh
        sum_Q_norm = sum_Q_mAh / nom_capacity_mAh
        d['EFC'].append(sum_Q_norm)
        d['E_cum_Wh'].append(sum_E_Wh)

        # only calc Q(V) at reference cycles (because voltage drops from different high discharge currents vary too much)
        Q_values = [0] * len(voltage_points)
        if referenceCycle:
            ref_indices.append(i)
            ret, Q_values = calc_Q_V(voltage, current_A, delta_t, voltage_points, nom_capacity_mAh, 8)
            soh = 100 * charge_mAh / nom_capacity_mAh
            if not ret:
                soh = 0
        else:
            soh = 0

        for feature, value in zip(Q_V_norm_features, Q_values):
            d[feature].append(value)

        #limit SOH to 100 %
        if soh > 100:
            soh = 100

        #battery51 second reference cycle correct SOH from 91 to 98.5 (runaway value)
        if (batt_label == 'battery51') and referenceCycle:
            if batt51_refCycle_cnt == 1:
                d['SOH'].append(98.5)
            else:
                d['SOH'].append(soh)
            batt51_refCycle_cnt += 1
        else: 
            d['SOH'].append(soh)

        #Calculate inner resistance
        meas_ok, tmp_res = return_inner_res_mOhm(mode, current_A, voltage, 2.0, 0.1)
        if meas_ok:
            res_mOhm = tmp_res 
        d['R_mOhm'].append(res_mOhm)

        # Do calculations for UV/OV
        # Do calculations for time at high/low voltage
        voltage_hist_cycle = calc_hist(voltage, delta_t, [6.2, 8.0])
        volt_hist += np.array(voltage_hist_cycle)
        d['sec_at_high_voltage'].append(volt_hist[-1])
        d['sec_at_low_voltage'].append(volt_hist[0])

        # Do calculations for integrals
        integrals['integral_high_volt_high_temp'] += integrate_two_feat_prod(delta_t, temp, 'gt', 40.0, voltage, 'gt', 8.0)
        integrals['integral_low_volt_high_temp'] += integrate_two_feat_prod(delta_t, temp, 'gt', 40.0, voltage, 'lt', 6.2)
        integrals['integral_low_volt_high_discharge'] += integrate_two_feat_prod(delta_t, voltage, 'lt', 6.2, current_C, 'gt', 1.0)
        integrals['integral_high_temp_high_discharge'] += integrate_two_feat_prod(delta_t, temp, 'gt', 40.0, current_C, 'gt', 1.0)
        integrals['integral_low_temp_high_discharge'] += integrate_two_feat_prod(delta_t, temp, 'lt', 0.0, current_C, 'gt', 1.0)

        C_squared = current_C ** 2
        integrals['integral_discharge_squared'] += (C_squared * delta_t).sum()
        integrals['integral_discharge_squared_T'] += (C_squared * (10.0 + abs(temp-20.0)) * delta_t).sum()

        for feature in integral_features:
            d[feature].append(integrals[feature])
       
        #Do temperature calculations here
        temp_hist_cycle = calc_hist(temp, delta_t, temp_bins)
        temp_hist += np.array(temp_hist_cycle)
        for feature, hist_value in zip(T_hist_features, temp_hist):
            d[feature].append(hist_value)
        
        avg_temp_cycle, avg_temp, integral_temp, sum_time = calc_average(temp, delta_t, integral_temp, sum_time)
        max_temp_cycle = temp.max()
        min_temp_cycle = temp.min() 
        max_temp = max(max_temp_cycle, max_temp)
        min_temp = min(min_temp_cycle, min_temp)
        temp_values = [avg_temp_cycle, max_temp_cycle, min_temp_cycle, avg_temp, max_temp, min_temp]
        for feature, value in zip(temp_features, temp_values):
            d[feature].append(value)

        temp_times = calc_thres_times(temp, delta_t, temp_thres)
        sum_time_temp_thres += np.array(temp_times)
        for feature, time in zip(temp_thres_features, sum_time_temp_thres):
            d[feature].append(time)

        #Do current calculations here  
        curr_hist_cycle = calc_hist(current_C, delta_t, curr_bins_C)
        curr_hist += np.array(curr_hist_cycle)

        discharge_delta_t = delta_t[mode == -1]
        discharge_current = current_C[mode == -1]
        avg_curr_cycle_C, avg_curr_C, integral_curr_C, sum_discharge_time =\
        calc_average(discharge_current, discharge_delta_t, integral_curr_C, sum_discharge_time)

        max_curr_cycle_C = current_C.max()
        max_curr_C = max(max_curr_cycle_C, max_curr_C)
        C_rate_values = [avg_curr_cycle_C, max_curr_cycle_C, avg_curr_C, max_curr_C]
        for feature, value in zip(C_rate_features, C_rate_values):
            d[feature].append(value)

        for feature, hist_value in zip(C_hist_features, curr_hist):
            d[feature].append(hist_value)
            
       
    # Make linear interpolation for SOH and Q(V)
    max_idx = max(ref_indices)
    min_idx = min(ref_indices)
    lin_int_features = ['SOH'] + Q_V_norm_features
       
    for i in range(0,len(indices)-1):
        if i > max_idx:
            for feature in lin_int_features:
                d[feature][i] = d[feature][max_idx]
        elif i < min_idx:
            for feature in lin_int_features:
                d[feature][i] = d[feature][min_idx]
        elif i not in ref_indices:
            for j,v in enumerate(ref_indices):
                if v > i:
                    # Found right index
                    x = 2*[0]
                    y = 2*[0]
                    last_ref_index = ref_indices[j-1]
                    x[0] = d['EFC'][last_ref_index]
                    x[1] = d['EFC'][v]
                    for feature in lin_int_features:
                        # make linear interpolation for every feature in lin_int_features
                        y[0] = d[feature][last_ref_index]
                        y[1] = d[feature][v]
                        x_interp = d['EFC'][i]
                        y_interp = np.interp(x_interp, x, y)     
                        d[feature][i] = y_interp
                    break

    # Initialize SOH history and SOH delta data
    soh_hist_features = ['SOH(-30EFC)', 'SOH(-60EFC)', 'SOH(-90EFC)', 'SOH(-120EFC)']
    soh_delta_features = ['delta_SOH(-30EFC)', 'delta_SOH(-60EFC)', 'delta_SOH(-90EFC)', 'delta_SOH(-120EFC)']
    soh_hist_offset = [30, 60, 90, 120]
    for feature in (soh_hist_features+soh_delta_features):
        d[feature] = []

    # Append SOH History data
    for num_efc in d['EFC']:
        for feature, offset in zip(soh_hist_features, soh_hist_offset):
            d[feature].append(get_soh(d['SOH'], d['EFC'], num_efc-offset))

    # Append delta SOH data
    for i, soh in enumerate(d['SOH']):
        for soh_delta_feature, soh_hist_feature in zip(soh_delta_features, soh_hist_features):
            delta_soh = d[soh_hist_feature][i] - soh
            d[soh_delta_feature].append(delta_soh)

    # Calculate RUL in EFC
    d['RUL_EFC'] = []
    for efc in d['EFC']:
        d['RUL_EFC'].append(d['EFC'][-1]-efc)

    df = pd.DataFrame(d)

    #drop rows until last row is reference cycle
    while df.iloc[-1]["isRefCycle"] == False:
        df = df.drop(len(df)-1)
    
    return df
````

````python
def create_cycle_data_frame(labels, base_path):
    complete_df = pd.DataFrame([])
    for label in labels:
        print(label)
        csv_name = label + ".csv"
        full_path = base_path + csv_name
        df_in = pd.read_csv(full_path)
        df_cleaned = clean_data(df_in)
        cycle_df = generate_cycle_based_data(label,df_cleaned)
        complete_df = pd.concat([complete_df,cycle_df],axis=0)
    return complete_df
````

# NASA randomized and recommissioned battery dataset

An accelerated Life Testing Dataset for Lithium-Ion Batteries with Constant and Variable Loading Conditions:

The dataset repository is organized into three main folders, each containing one group of life cycled battery packs. 
Within each folder individual battery packs own their dedicated csv file for continuous data logging, which are named with their respective battery pack number.

The folders are named: 

- regular_alt_batteries: Containing one csv file for each battery pack cycled at the same load level or load range throughout lifetime
- recommissioned_batteries: Containing one csv file for each battery pack cycled at different load levels at varying life stages
- second_life_batteries: Containing one csv file for each second life battery pack cycled at constant current througout the second life 

The columns in each csv file contain the following data with the provided units: 

The following columns contain data throughout the cycling process: 

- start time: [mm:dd:yyyy hh:mm:ss] (start time of each cycle day for each battery pack, in most cases approximately 24h)
- relative time: [s] (the relative time is continuous from the beginning of the entire life cycle to the failure of each battery pack)
- mode: -1 = discharge, 0 = rest, 1 = charge
- voltage charger: [V] (continuous battery pack voltage measurement right after the connection of the battery pack to the charger board)
- temperature battery: [C] (continuous temperature measurement on the battery cell electrode surface)

Those columns only contain data when the respective battery is connected to the load board and performing a discharge mission

- voltage load: [V] (battery pack voltage measured on the load board)
- current load: [A] (discharge current measured on the load board using current sense resistors)
- temperature mosfet: [C] (temperature on the load board mosfets, measured for safety purposes)
- temperature resistor: [C] (temperature on the load board current sense resistor, measured for safety purposes)
- mission type: 0 = reference discharge (constant current at 2.5A), 1 = regular mission

## Regular accelerated life test (regular alt batteries folder):

This folder contains the data files from battery packs subject to constant and variable loading conditions, 
where the current range of the latter remains the same throughout the battery life.

The assigned battery packs in this folder are numbered as
follows:

– Constant current:
Those battery packs were subjected to the same load level during discharge missions over their respective life time.

The here provided averaged current levels are approximated, and might vary slightly:

* 9.30A: Battery pack 0.1 and 1.1
* 12.9A: Battery pack 3.1 and 2.2
* 14.3A: Battery pack 2.3 and 5.2
* 16.0A: Battery pack 0.0 and 1.0
* 19.0A: Battery pack 2.0, 3.0 and 2.1

– Variable current:

Those battery packs are subjected to variable current within a discharge cycle.
The current level segments are kept constant for either 40, 60 or 80 seconds before switching to another level.
Both, the amplitude and length of each current level segments was randomly selected to create a pool of different random loading discharge missions.

The here provided averaged current levels are approximated, and might vary slightly:

* 14.3A average (12.9A - 16A range): Battery pack 4.1 and 5.1
* 17.0A average (16A - 19A range): Battery pack 4.0 and 5.0


## Recommissioned batteries (recommissioned batteries folder):

This folder contains the data files from battery packs where the constant loading conditions were changed throughout the battery life. 
The assigned battery packs in this folder are numbered as follows:

– Two life stages:
The following batteries were subjected to constant current levels at the first provided amplitudes until reaching an capacity level of approximately 2.2Ah,
and then switched to the second current level until reaching end of life

The here provided averaged current levels are approximated, and might vary slightly:

* 16A, 14.3A: Battery pack 3.2
* 14.3A, 16A: Battery pack 5.3
* 16A, 12.9A: Battery pack 0.2 and 3.3
* 16A, 9.30A: Battery pack 1.2 and 2.4

– Three life stages:
Those batteries were subjected to current level switches at approximately 2.2Ah and 2.0Ah of capacity.
The here provided averaged current levels are approximated, and might vary slightly:

* 16A, 14.3A, 7.5A: Battery pack 0.3
* 16A, 7.5A, 14.3A: Battery pack 2.5


## Second life batteries (second life batteries folder):

This folder contains the data files from battery packs that consist of battery cells survived from previously cycled battery packs. 
The assigned battery packs in this folder are numbered as follows:

– 2nd life battery packs:
The following batteries were subjected to constant current levels at the first provided amplitudes until reaching end of life (EOL).
Survivor cells were then reused and subjected to the second provided current level until reaching EOL

The here provided averaged current levels are approximated, and might vary slightly:

* 16.0A, 9.3A: Battery pack 1.3 (First life battery packs: 0.0 & 1.1)
* 14.3A, 7.5A: Battery pack 3.6 (First life battery packs: 2.3 & 5.2)
* 12.9A, 5.0A: Battery pack 5.4 (First life battery packs: 2.2 & 3.1)

````python
#Clean data and drop not used columns
def clean_data(df_in):
    df = df_in.copy()
    df['temperature_battery'] = df['temperature_battery'].astype(float)
    
    mean_temp = df['temperature_battery'].mean()
    df['temperature_battery'] = df['temperature_battery'].fillna(mean_temp)
    
    def replace_negative_temp(x):
        if x <= 0:
            return mean_temp
        else:
            return x
    
    df['temperature_battery'] = df['temperature_battery'].apply(replace_negative_temp)
    df['current_load'] = df['current_load'].fillna(0)
    df = df.drop(columns=['start_time','temperature_mosfet','temperature_resistor','voltage_load'])

    #df.loc[1:,'delta_t'] = df['time']

    time_array = np.array(df['time'])
    delta_t_array = time_array[1:] - time_array[:-1]
    delta_t_list = list(delta_t_array)
    delta_t_list = [0] + delta_t_list
    delta_t_array = np.array(delta_t_list)
    df['delta_t'] = pd.Series(delta_t_array)
    
    return df

df = clean_data(df_in)

df.head(3)
````

````python
def calc_Q_V(data,voltages,nom_cap_mAh,min_start_voltage):
    '''
    Calculates the normalized discharge from full charge for given voltage points
    '''
    Q_norm = []
    if data.iloc[0]['voltage_charger'] < min_start_voltage:
        return False, []
    i=0
    for idx in range(len(voltages)):
        voltage = voltages[idx]
        while i < len(data)-1 and data.iloc[i]['voltage_charger'] > voltage:        
            i += 1
        q = (data.iloc[:i]['current_load'] * data.iloc[:i]['delta_t']).sum()
        Q_norm.append(q/3.6/nom_cap_mAh)
    return True, Q_norm

def return_inner_res_mOhm(data,min_curr,min_volt_drop):
    '''
    Calculates the inner resistance when the mode changes from discharge to idle
    '''
    i=0
    inner_resistance = 0
    last_idx = 0
    last_res = 0
    
    #check when mode goes into discharge
    while i < len(data) and data.iloc[i]['mode'] != -1:
        i += 1
        
    #check when mode goes into idle
    while i < len(data) and data.iloc[i]['mode'] != 0:
        i += 1
        
    #only calc inner resistance when discharge current is over a threshold
    current = data.iloc[i-1]['current_load']
    meas_OK = False
    if current > min_curr:
        voltage_discharge = data.iloc[i-1]['voltage_charger']
        #check when voltage increases a threshold to get the ohmic voltage drop
        #and no relaxation voltage drop
        offset=0
        while offset < 10:
            voltage = data.iloc[i+offset]['voltage_charger']
            if voltage > (voltage_discharge + min_volt_drop):
                voltage_open = voltage
                inner_resistance = 1000*(voltage_open - voltage_discharge)/current
                meas_OK = True
                break
            offset += 1
    return meas_OK, inner_resistance

def isReferenceCycle(data):
    '''
    checks if given cycle is a reference cycle with 2.5A discharge
    '''
    for i in range(len(data)):
        if (data.iloc[i]['mission_type'] == 0) and (2 < data.iloc[i]['current_load'] < 3):
            return True
    return False

def calc_charge(data):
    '''
    calculates the full charge throuput for the given data
    '''
    charge_As = (data['current_load']*data['delta_t']).sum()
    charge_mAh = charge_As / 3.6
    return charge_mAh

def calc_energy(data):
    '''
    calculates the full energy throuput for the given data
    '''
    energy_Ws = (data['current_load']*data['voltage_charger']*data['delta_t']).sum()
    energy_Wh = energy_Ws/3600
    return energy_Wh

def calc_hist(feature,data,bins):
    '''
    Calculates histogram and avg,max,min for a given feature
    '''
    hist = [0]*(len(bins)+1)

    maxi = data[feature].max()
    avg = (data[feature] * data['delta_t']).sum() / data['delta_t'].sum()
    mini = data[feature].min()

    for i in range(1,len(bins)):
        tmp = data[(data[feature]<bins[i]) & (data[feature]>=bins[i-1])]  
        hist[i] = tmp['delta_t'].sum()

    tmp = data[data[feature]<bins[0]]
    hist[0] = tmp['delta_t'].sum()

    tmp = data[data[feature]>bins[-1]]
    hist[-1] = tmp['delta_t'].sum()
    
    return hist, maxi, avg, mini


def return_full_charged_cycle_indices(data, charged_volt, discharged_volt):
    '''
    return all indices where a new cycle starts
    '''
    indices = []
    max_idx = len(data) - 1
    voltage = 'voltage_charger'
    i=0
    
    # check time when battery is fully charged
    while i < max_idx:
        if data['mode'][i]==0 and data[voltage][i] >= charged_volt:
            break
        i += 1
    
    while i < max_idx:
        indices.append(i)
        # wait until battery is discharging
        while (i < len(data)) and (data['mode'][i] != -1):
            i += 1
        
        # check time when battery is fully discharged
        charged = False
        while i < max_idx:
            if data['mode'][i]==0 and data[voltage][i] <= discharged_volt:
                break         
            i += 1

        # check time when battery is fully charged
        while i < max_idx:
            if data['mode'][i]==0 and data[voltage][i] >= charged_volt:
                break
            i += 1
            
    return indices
    

def generate_cycle_based_data(batt_label, data):
    """
    Generate cycle based data from the time based full cycle battery test data
    """
    d = {}
    d['batt_label'] = []
    d['isRefCycle'] = []
    d['time_sec'] = []

    d['Q_cum_norm'] = []
    d['E_cum_norm'] = []

    d['T_avg'] = []
    d['T_max'] = []
    d['T_min'] = []
    d['perc_T_st_-20'] = []
    d['perc_T_-20_0'] = []
    d['perc_T_0_20'] = []
    d['perc_T_20_40'] = []
    d['perc_T_40_60'] = []
    d['perc_T_60_80'] = []
    d['perc_T_80_100'] = []
    d['perc_T_gt_100'] = []
    
    d['C_rate_avg'] = []
    d['C_rate_max'] = []
    d['perc_C_rate_0_0.5'] = []
    d['perc_C_rate_0.5_1'] = []
    d['perc_C_rate_1_2'] = []
    d['perc_C_rate_2_3'] = []
    d['perc_C_rate_3_4'] = []
    d['perc_C_rate_4_6'] = []
    d['perc_C_rate_6_8'] = []
    d['perc_C_rate_gt_8'] = []

    d['Q_norm_at_3.8_V'] = []
    d['Q_norm_at_3.6_V'] = []
    d['Q_norm_at_3.4_V'] = []
    d['Q_norm_at_3.2_V'] = []

    d['R_norm'] = []
    d['SOH'] = []

    temp_bins = [-20,0,20,40,60,80,100]
    temp_hist = np.array(8*[0.0])
    temp_hist_perc = np.array(8*[0.0])
    sum_temp = 0
    avg_temp = 0
    max_temp = -1000
    min_temp = 1000

    curr_bins_C = np.array([0.5,1,2,3,4,6,8])
    curr_hist = np.array(8*[0.0])
    curr_hist_perc = np.array(8*[0.0])
    sum_curr = 0
    avg_curr = 0
    max_curr = -1000
    min_curr = 1000

    voltage_points = [7.6,7.2,6.8,6.4]
    Q_values = [0] * len(voltage_points)

    sum_Q_mAh = 0
    sum_E_Wh = 0

    inner_res_norm = 1
    first_res_meas = False
    nom_res_mOhm = 40

    indices = return_full_charged_cycle_indices(data, 8, 6.3)
    #print(indices)

    ref_indices = []
    first_idx = indices[0]
    
    for i in range(0,len(indices)-1):
        d['batt_label'].append(batt_label)
        start_idx = indices[i]
        end_idx = indices[i+1]
        
        cycle = data.loc[start_idx:end_idx]
        referenceCycle = isReferenceCycle(cycle)
        d['isRefCycle'].append(referenceCycle)
        d['time_sec'].append(cycle.iloc[-1]['time'])

        charge_mAh = calc_charge(cycle)
        energy_Wh = calc_energy(cycle)

        sum_Q_mAh += charge_mAh
        sum_E_Wh += energy_Wh

        # Calculate nom_capacity_mAh when first cycle
        if i == 0:
            nom_capacity_mAh = charge_mAh
            curr_bins_A = curr_bins_C * nom_capacity_mAh / 1000
            nom_Energy_Wh = energy_Wh

        if referenceCycle:
            ref_indices.append(i)
            ret, Q_values = calc_Q_V(cycle,voltage_points,nom_capacity_mAh,8)
            if not ret:
                Q_values = [0] * len(voltage_points)
        else:
            Q_values = [0] * len(voltage_points)

        d['Q_norm_at_3.8_V'].append(Q_values[0])
        d['Q_norm_at_3.6_V'].append(Q_values[1])
        d['Q_norm_at_3.4_V'].append(Q_values[2])
        d['Q_norm_at_3.2_V'].append(Q_values[3])

        sum_Q_norm = sum_Q_mAh / nom_capacity_mAh
        sum_E_norm = sum_E_Wh / nom_Energy_Wh

        d['Q_cum_norm'].append(sum_Q_norm)
        d['E_cum_norm'].append(sum_E_norm)

        soh = 100*charge_mAh/nom_capacity_mAh
        d['SOH'].append(soh)

        #Calculate inner resistance
        meas_ok, tmp_res = return_inner_res_mOhm(cycle,8,0.1)

        if meas_ok:
            inner_res_mOhm = tmp_res
            if not first_res_meas:
                first_res_meas = True
                nom_res_mOhm = inner_res_mOhm
            else:
                inner_res_norm = inner_res_mOhm / nom_res_mOhm

        d['R_norm'].append(inner_res_norm)
        
        #print(soh)
        #print(sum_Q_norm)
        #print(sum_E_norm)
            
        #print(nom_capacity_mAh)
        #print(curr_bins_A)
        
        #Do temperature calculations here
        temp_hist_cycle, max_temp_cycle, avg_temp_cycle, min_temp_cycle = calc_hist('temperature_battery',cycle,temp_bins)
        temp_hist += np.array(temp_hist_cycle)
        temp_hist_perc = 100 * temp_hist / temp_hist.sum()
        sum_temp += avg_temp_cycle * (data['time'][end_idx] - data['time'][start_idx])
        avg_temp = sum_temp / (data['time'][end_idx] - data['time'][first_idx])
        if max_temp_cycle > max_temp:
            max_temp = max_temp_cycle
        if min_temp_cycle < min_temp:
            min_temp = min_temp_cycle

        d['T_avg'].append(avg_temp)
        d['T_max'].append(max_temp)
        d['T_min'].append(min_temp)
        d['perc_T_st_-20'].append(temp_hist_perc[0])
        d['perc_T_-20_0'].append(temp_hist_perc[1])
        d['perc_T_0_20'].append(temp_hist_perc[2])
        d['perc_T_20_40'].append(temp_hist_perc[3])
        d['perc_T_40_60'].append(temp_hist_perc[4])
        d['perc_T_60_80'].append(temp_hist_perc[5])
        d['perc_T_80_100'].append(temp_hist_perc[6])
        d['perc_T_gt_100'].append(temp_hist_perc[7])

        #print(temp_hist_perc)
        #print(max_temp)
        #print(avg_temp)
        #print(start_idx)
        
        #Do current calculations here, ToDo: Calculate real currents with C values
        
        curr_hist_cycle, max_curr_cycle, avg_curr_cycle, min_curr_cycle = calc_hist('current_load',cycle,curr_bins_A)
        curr_hist += np.array(curr_hist_cycle)
        curr_hist_perc = 100 * curr_hist / curr_hist.sum()
        sum_curr += avg_curr_cycle * (data['time'][end_idx] - data['time'][start_idx])
        avg_curr_C = sum_curr / (data['time'][end_idx] - data['time'][first_idx]) * 1000 / nom_capacity_mAh
        if max_curr_cycle > max_curr:
            max_curr = max_curr_cycle
            max_curr_C = 1000 * max_curr / nom_capacity_mAh

        d['C_rate_avg'].append(avg_curr_C)
        d['C_rate_max'].append(max_curr_C)
        d['perc_C_rate_0_0.5'].append(curr_hist_perc[0])
        d['perc_C_rate_0.5_1'].append(curr_hist_perc[1])
        d['perc_C_rate_1_2'].append(curr_hist_perc[2])
        d['perc_C_rate_2_3'].append(curr_hist_perc[3])
        d['perc_C_rate_3_4'].append(curr_hist_perc[4])
        d['perc_C_rate_4_6'].append(curr_hist_perc[5])
        d['perc_C_rate_6_8'].append(curr_hist_perc[6])
        d['perc_C_rate_gt_8'].append(curr_hist_perc[7])

    
        #print(curr_hist_perc)
        #print(max_curr_C)
        #print(avg_curr_C)

    
    # Make linear interpolation for Q(V)
    max_idx = max(ref_indices)
    min_idx = min(ref_indices)

    feat_names = ['Q_norm_at_3.8_V', 'Q_norm_at_3.6_V', 'Q_norm_at_3.4_V', 'Q_norm_at_3.2_V']
        
    for i in range(0,len(indices)-1):
        if i > max_idx:
            for feat_name in feat_names:
                d[feat_name][i] = d[feat_name][max_idx]
        elif i < min_idx:
            for feat_name in feat_names:
                d[feat_name][i] = d[feat_name][min_idx]
        elif i not in ref_indices:
            for j,v in enumerate(ref_indices):
                if v > i:
                    # Found right index
                    x = 2*[0]
                    y = 2*[0]
                    x[0] = d['Q_cum_norm'][ref_indices[j-1]]
                    x[1] = d['Q_cum_norm'][v]
                    for feat_name in feat_names:
                        y[0] = d[feat_name][ref_indices[j-1]]
                        y[1] = d[feat_name][v]
                        x_interp = d['Q_cum_norm'][i]
                        y_interp = np.interp(x_interp, x, y)     
                        d[feat_name][i] = y_interp
                    break
        
        
    ret = pd.DataFrame(d)
    return ret

cycle_based_df = generate_cycle_based_data("battery20",df)
cycle_based_df.to_csv("cycles_battery20.csv")
cycle_based_df
````
